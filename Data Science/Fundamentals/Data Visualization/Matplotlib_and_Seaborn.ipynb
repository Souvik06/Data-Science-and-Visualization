Why Data Visualization?

------

- Using easy-to-digest visualizations, information can be quickly shared.
- Find hidden patterns and relationships in data.
- It's easy to track trends like sales, shares and so on.
- Storytelling with data creates more impactful results.

Let's group the data by Country and assign the Year as the index

```python
GermanyHealthExp = healthexp.groupby('Country').get_group('Germany').drop(columns = 'Country').set_index('Year')
FranceHealthExp = healthexp.groupby('Country').get_group('France').drop(columns = 'Country').set_index('Year')
BritainHealthExp = healthexp.groupby('Country').get_group('Great Britain').drop(columns = 'Country').set_index('Year')
JapanHealthExp = healthexp.groupby('Country').get_group('Japan').drop(columns = 'Country').set_index('Year')
USAHealthExp = healthexp.groupby('Country').get_group('USA').drop(columns = 'Country').set_index('Year')
CanadaHealthExp = healthexp.groupby('Country').get_group('Canada').drop(columns = 'Country').set_index('Year')
```

# Stacked Bar

```python
plt.bar(GermanyHealthExp.loc[2000:2020].index,GermanyHealthExp.loc[2000:2020]['Life_Expectancy'])
plt.bar(FranceHealthExp.loc[2000:2020].index,FranceHealthExp.loc[2000:2020]['Life_Expectancy'],
        bottom = GermanyHealthExp.loc[2000:2020]['Life_Expectancy'])
plt.bar(BritainHealthExp.loc[2000:2020].index,BritainHealthExp.loc[2000:2020]['Life_Expectancy'],
        bottom = GermanyHealthExp.loc[2000:2020]['Life_Expectancy'] + FranceHealthExp.loc[2000:2020]['Life_Expectancy'])
plt.bar(JapanHealthExp.loc[2000:2020].index,JapanHealthExp.loc[2000:2020]['Life_Expectancy'],
        bottom = GermanyHealthExp.loc[2000:2020]['Life_Expectancy'] + FranceHealthExp.loc[2000:2020]['Life_Expectancy'] + BritainHealthExp.loc[2000:2020]['Life_Expectancy'])
plt.bar(USAHealthExp.loc[2000:2020].index,USAHealthExp.loc[2000:2020]['Life_Expectancy'],
        bottom = GermanyHealthExp.loc[2000:2020]['Life_Expectancy'] + FranceHealthExp.loc[2000:2020]['Life_Expectancy'] + BritainHealthExp.loc[2000:2020]['Life_Expectancy'] + JapanHealthExp.loc[2000:2020]['Life_Expectancy'])
plt.bar(CanadaHealthExp.loc[2000:2020].index,CanadaHealthExp.loc[2000:2020]['Life_Expectancy'],
        bottom = GermanyHealthExp.loc[2000:2020]['Life_Expectancy'] + FranceHealthExp.loc[2000:2020]['Life_Expectancy'] + BritainHealthExp.loc[2000:2020]['Life_Expectancy'] + JapanHealthExp.loc[2000:2020]['Life_Expectancy'] + USAHealthExp.loc[2000:2020]['Life_Expectancy'])
plt.legend(['Germany', 'France', 'Great Britain', 'Japan', 'USA', 'Canada'],loc=0,bbox_to_anchor=(1,1))
plt.xlabel("Year")
plt.ylabel("Life_Expectancy")
plt.show()
```

```python
plt.barh(GermanyHealthExp.loc[2000:2020].index,GermanyHealthExp.loc[2000:2020]['Spending_USD'])
plt.barh(FranceHealthExp.loc[2000:2020].index,FranceHealthExp.loc[2000:2020]['Spending_USD'],
        left = GermanyHealthExp.loc[2000:2020]['Spending_USD'])
plt.barh(BritainHealthExp.loc[2000:2020].index,BritainHealthExp.loc[2000:2020]['Spending_USD'],
        left = GermanyHealthExp.loc[2000:2020]['Spending_USD'] + FranceHealthExp.loc[2000:2020]['Spending_USD'])
plt.barh(JapanHealthExp.loc[2000:2020].index,JapanHealthExp.loc[2000:2020]['Spending_USD'],
        left = GermanyHealthExp.loc[2000:2020]['Spending_USD'] + FranceHealthExp.loc[2000:2020]['Spending_USD'] + BritainHealthExp.loc[2000:2020]['Spending_USD'])
plt.barh(USAHealthExp.loc[2000:2020].index,USAHealthExp.loc[2000:2020]['Spending_USD'],
        left = GermanyHealthExp.loc[2000:2020]['Spending_USD'] + FranceHealthExp.loc[2000:2020]['Spending_USD'] + BritainHealthExp.loc[2000:2020]['Spending_USD'] + JapanHealthExp.loc[2000:2020]['Spending_USD'])
plt.barh(CanadaHealthExp.loc[2000:2020].index,CanadaHealthExp.loc[2000:2020]['Spending_USD'],
        left = GermanyHealthExp.loc[2000:2020]['Spending_USD'] + FranceHealthExp.loc[2000:2020]['Spending_USD'] + BritainHealthExp.loc[2000:2020]['Spending_USD'] + JapanHealthExp.loc[2000:2020]['Spending_USD'] + USAHealthExp.loc[2000:2020]['Spending_USD'])
plt.legend(['Germany', 'France', 'Great Britain', 'Japan', 'USA', 'Canada'],loc=0,bbox_to_anchor=(1,1))
plt.ylabel("Year")
plt.xlabel("Spending_USD")
plt.show()
```

```python
healthDict = {}
for group,data in healthexpByYear.groupby('Country'):
    healthDict[group]=data['Life_Expectancy']
```

```python
heathExpPerCountry = pd.DataFrame(healthDict)
heathExpPerCountry.head()
```

```python
heathExpPerCountry.dropna(inplace= True)
heathExpPerCountry.head()
```

```python
heathExpPerCountry.plot(kind='bar',stacked=True)
plt.show()
```

# Pie Plot

```python
flights.head()
```

```python
data_label = list(set(flights['year']))
data_count = []
for group, data in flights.groupby('year'):
    data_count.append(data['passengers'].mean())
plt.pie(data_count,labels = data_label,autopct = '%0.1f')
plt.title('Average Passengers % per year')
plt.show()
```
